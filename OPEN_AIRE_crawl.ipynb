```python
links_to_crawl = []

for item in range(0, len(list_with_pages)):

    r = requests.get(list_with_pages[item])
    soup = BeautifulSoup(r.text, "lxml")
    
    a_href = soup.findAll("a", {"class":"uk-link uk-text-decoration-none uk-width-expand"})

    print(a_href[0].get("href"))
    print(len(a_href))
    # links_list = []

    for i in range(len(a_href)):
        print ("Found the URL:", a_href[i].get("href"))
        if a_href[i].get("href") not in links_to_crawl:
            links_to_crawl.append(a_href[i].get("href"))
    
    # links_to_crawl.append(links_list[0:50])

print(len(links_to_crawl))
print(len(set(links_to_crawl)))
```

```python
# links_50 = links_list[0:50]
site = "https://explore.openaire.eu"
abstracts_list = []
count = 0

for link in links_to_crawl:
    link = site + link

    r = requests.get(link)
    print(r)
    soup = BeautifulSoup(r.text, "lxml")
    
    buffer_dic = {}
    
    try:
        buffer_dic["abstract"] = soup.find("div", {"class":"multi-line-ellipsis lines-10"}).get_text()
    except AttributeError:
        continue

    buffer_dic["title"] = soup.find("div", {"class":"uk-text-break multi-line-ellipsis lines-5"}).get_text() 
    buffer_dic["doi"] = soup.find("span", {"id": "doi"}).get_text()

    metrics_2 = soup.find_all("div", {"class": "uk-flex uk-flex-wrap landing-metrics-overview"})
    text = ""
    for j in metrics_2:
        text = text + j.get_text()
    
    buffer_dic["metrics"] = text
    abstracts_list.append(buffer_dic)
    count += 1
    print(count)

print(len(abstracts_list))
print(type(abstracts_list[0]))
print(abstracts_list[0])
```
